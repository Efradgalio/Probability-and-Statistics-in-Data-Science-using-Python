## 3.1 Counting

### Set Size 
*The number of elements in a set S is called its size, or cardinality, and denotes |S| or #S.*
<t> Bits --> |{0,1}| = 2
<t> Coin --> |{heads,tails}| = 2
<t> Empty set --> |$\varnothing$| = 0
<t> Integers --> |$\mathbb{Z}$| = |$\mathbb{N}$| = |$\mathbb{P}$| = $\infty$

### Integer Intervals
|{m,...,n}| = n-m+1
<t> |{5,...,5}| = 1 = 5-5 + 1
<t> |{1,2,3}| = 3 = 3-1+1

### Integer Multiples
(n] = [n] = {1,...,n}
<t> ${}_d(n] = \{ 1 \leq i \leq n : d \mid i \} $
<t> ${}_3(8]$  = {3,6} = {1x3, 2x3}
<t> ${}_3(9]$  = {3,6,9} = {1x3, 2x3, 3x3}
<t> $\mid {}_d(n] \mid $ = |n/d|
<t> $\mid {}_3(8] \mid $ = |8/3| = 2
<t> $\mid {}_3(9] \mid $ = |9/3| = 3

## 3.2 Disjoint Unions

*A union of disjoint sets is called a disjoint union*
<t> $\mid A \cup B \mid = \mid A \mid + \mid B \mid $ 
<t> We can see for disjoint sets, the size of union is the sum of the sizes --> Addition Rule, it has numerous applications & implications
<t> Reason U $\approx$ +

### Example
Class has 2 boys and 3 girls, # Students = ?
<t> This is disjoint unions case, so we can say :  $\mid Boys \cup Girls \mid = \mid Boys \mid + \mid Girls \mid $ = 2 + 3 = 5

### Complements
$ A \cup A^c = \Omega$
<t> $ \mid \Omega \mid \; = \; \mid  A \cup A^c \mid \; =\; \mid A \mid + \mid A^c \mid$
<t> $\mid A^c \mid \; = \; \mid \Omega \mid - \mid A \mid$ --> Subtraction (or complement) Rule
<t> Reason set difference $\approx$ -

### Think Outside the Circle
$\mid A^c \mid \; = \; \mid \Omega \mid - \mid A \mid$  --> Can be handy for large or complex sets !
<t> $\mid A \mid \; = \; \mid \Omega \mid - \mid A^c \mid$
<t> Example :
<t> $ A = \{ i \in [100] : 3 \nmid i\} = \{1,2,4,5,7,..,100\}$
<t> $\Omega = \{1,...,100\}$
<t> $A^c = \{i \in [100] : 3 \mid i\} = \{3,6,9,...,99\}$ --> $\mid A^c \mid = 33$
<t> $\mid A \mid$ = $\mid \Omega \mid - \mid A^c \mid = 100 - 33 = 67$


### General Subtraction Rule
$ A \subseteq B $ --> $ B = A \cup (B-A)$
<t> $\mid B \mid \; = \; \mid A \mid + \mid B-A \mid$
<t> $\mid B - A \mid \; = \; \mid B \mid - \mid A \mid$ --> Subtraction Rule
    

## 3.3 General Unions

In General : $\mid A \cup B \mid \; \neq \; \mid A \mid + \mid B \mid$
<t> Example : |{a} u {a}| = |{a}| = 1 (not 2)
<t> General Unions : $\mid A \cup B \mid \; = \; \mid A \mid + \mid B \mid - \mid A \cap B \mid$ -->  Called Principle of Inclusion-Exclusion(PIE)

### Divisibility by 2 Numbers
$D_{2\vee 3} = \{i \in [100] : 2\mid i \vee 3 \mid i \} = \{2,3,4,6,8,...,100\}$ --> $ \mid D_{2\vee 3}\mid$ = ?

<t> $D_{2} = \{i \in [100] : 2 \mid i \} = \{2,4,6,...,100\}$ --> Size is 50
<t> $D_{3} = \{i \in [100] : 2 \mid i \} = \{3,6,9,...,99\}$ --> Size is 33
<t> $D_{2} \cap D_{3}= \{i \in [100] : 2\mid i \wedge 3 \mid i \} = \{i \in [1000] : 6 \mid i \}$ --> Size is 16  
<t> $ \mid D_{2\vee 3}\mid \; = \; \mid D_{2} \mid + \mid D_{3} \mid - \mid D_{2} \cap D_{3} \mid$ = 50 + 33 - 16 = 67

### Multiple Sets
Three Sets : $ \mid A \cup B \cup C \mid \; = \; \mid A \mid + \mid B \mid +  \mid C \mid - \mid A \cap B \mid - \mid A \cap C \mid - \mid B\cap C \mid + \mid A \cap B \cap C \mid$

## 3.4 Cartesian Powers

### Analogies between number and set operations

### Cartesian Powers of a set
Cartesian product of a set with itself is a Cartesian Power.
<t> $A^2 = A \; x \; A $ --> Cartesian square
<t> $A^n = A \; x \; A \; x ... \; x \; A $ --> n'th Cartesian Power
<t> $\mid A^n \mid \; = \mid A \mid \; x \; \mid A \mid \; x ... \; x \; \mid A \mid \; = \; \mid A \mid^n$ --> n'th Cartesian Power

### Binary Strings
$\{0,1\}^n = \{ length-n binary strings\}$

  n            | Sets                 | Strings          | Size
-------------  | ---------------      | -------------    |  -------------
0              |  $\{0,1\}^0$         | $\varnothing$    | 1
1              |  $\{0,1\}^1$         | 0,1              | 2
2              |  $\{0,1\}^2$         | 00,01,10,11      | 4
...            | ...                  |   ...            | ...
n              | $\{0,1\}^n$          |  0...0,...,1...1 | $2^n$

$\mid \{0,1\}^n \mid \; = \; \mid \{0,1\} \mid^n  = 2^n$ 


### Subsets
The power set of S, denoted $\mathbb{P}(S)$, is the collection of all subsets of S
<t> $ \mathbb{P}(\{a,b\}) = \{\{\}, \{a\}, \{b\}, \{a,b\}\}$
<t> $ \mid \mathbb{P}(S) \mid \; = \; \mid \{0,1\}^{\mid S \mid} \mid \; = \; 2^{\mid S \mid}$
<t> The size of the power set is the power of the set size

### Functions
A function from A to B maps every element $a \in A$ to an element $f(a) \in B$
<t> $ \{ \; functions \; from \; \{1,2,3\} \; to \; \{p,u\} \} \; = \; \{p,u\} \times \{p,u\} \times \{p,u\}$
<t> $ \{ \; functions \; from \; \{1,2,3\} \; to \; \{p,u\} \}$  = $2 \times 2 \times 2$ = $2^3$ = $\mid \{p,u\} \mid^{\mid \{1,2,3\} \mid} $
    
<t> # functions from A to B = $\mid B^{\mid A \mid} \mid \; = \; \mid B \mid^{\mid A \mid}$

## 3.6 Variations

### Find a natural counting question whose answer is a double exponential
$\mid \mathbb{P} (\mathbb{P}(S)) \mid$ = $2^{\mid \mathbb{P}(S) \mid}$ = $2^{2^{\mid S \mid}}$
<t> Functions from $\{0,1\}^n$ to $\{0,1\}$ --> $\{0,1\}^{\{0,1\}^n}$ = $2^{2^{n}}$

Combine, Addition, Subtraction, Multiplication --> Determine size of more complex

### Example PIN
Personal Identification Number --> 2174 
<t> # 4-digit PIN's = ?
<t> D = {0,...,9}
<t> $\mid D \mid^4$ = 10.000

Number of 3-5 digit PIN's --> 314 , 2246 , 79380
<t> D = {0,...,9}
<t> $\{PIN's\}$ = $D^3 \cup D^4 \cup D^5$

<t> Allowed --> 10,000 - 17 = 9083

### PINs containing Zero
Number of PIN's containing zero --> 8093 can , but 2534 cannot !
<t> D = {0,1,...,9}
<t> Z = {0} --> $Z^c$ = {1,...,9}

#### Using 2-digits: Inclusion-Exclusion
Z1 --> 00,03,05 can , but 50,55 cannot ! --> # = 10
<t> Z2 --> 00,30,50 can , but 05,55 cannot ! --> # = 10

<t> $\mid \exists Z \mid$ = $\mid Z1 \mid + \mid Z2 \mid - \mid Z1 \cap Z2 \mid$ --> 10 + 10 - 1 = 19


#### Using 2-digits: Complement Rule
$\Omega = D^2$
<t> $\mid \exists Z \mid^c$ = 9 x 9 = 81 
<t> $\mid \exists Z \mid$ = $ \mid D \mid^2 - \mid \exists Z \mid^c$ = 100 - 81 = 19
    
#### n-digits: Inclusion-Exclusion vs Complement Rule
1. Inclusion-Exclusion --> Very Hard when n larger
2. Complement Rule --> Easy when n larger
